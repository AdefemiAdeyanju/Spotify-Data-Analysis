```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
tracks = pd.read_csv('C:\\Users\\adefe\\Documents\\DataScienceProjects\\OnlineProjects\\Spotify Data Analysis\\tracks.csv')
tracks.head()
```

```python
#Null values
pd.isnull(tracks).sum()
```

```python
tracks.describe()
```

```python
#Basic information
tracks.info()
```

```python
#Least popular songs
sorted_tracks = tracks.sort_values('popularity', ascending = True).head(10)
sorted_tracks
```

```python
#Transpose of describe method
tracks.describe().transpose()
```

```python
#Most popular songs
most_popular = tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
#Set release date as index
tracks.set_index('release_date', inplace = True)
tracks.index = pd.to_datetime(tracks.index)
tracks.head()
```

```python
#Locate Artiste
tracks[['artists']].iloc[18]
```

```python
#Convert duration to miliseconds
tracks["duration"]= tracks["duration_ms"].apply(lambda x: round(x/1000))
tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
tracks.duration.head()
```

```python
sns.set()
corr = tracks.drop(['key','mode','explicit'], axis= 1).corr(method= 'pearson')
plt.figure(figsize=(14,6))
heatmap= sns.heatmap(corr, annot= True, fmt= '.1g', vmin= -1, vmax= 1, center= 0, cmap= 'inferno', linewidths = 1, linecolor= 'Black')
heatmap.set_title('Correlation Heatmap Between Variable')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
plt.show()
```

```python
#Sampling Data
sample = tracks.sample(int(0.004*len(tracks)))
(len(sample))
```

```python
#Regression Plot
plt.figure(figsize=(10,6))
sns.regplot(data = sample, y = 'loudness', x = 'energy', color= 'c').set(title= 'Loudness vs Energy Correlation')
plt.show()
```

```python
#Regression Plot
sns.set()
plt.figure(figsize=(10,6))
sns.regplot(data = sample, y = 'popularity', x = 'acousticness', color= 'b').set(title= 'Popularity vs Acousticness')
plt.show()
```

```python
#Year Column
tracks['dates']= tracks.index.get_level_values('release_date')
tracks.dates= pd.to_datetime(tracks.dates)
years= tracks.dates.dt.year
```

```python
sns.displot(years, discrete= True, aspect = 2, height = 5, kind= 'hist').set(title= 'Number of songs per year')
plt.show()
```

```python
#total duration bar plot
total_dr = tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x= years, y = total_dr, ax=ax, errwidth = False).set(title='Year vs Duration')
plt.xticks(rotation = 90)
plt.show()
```
